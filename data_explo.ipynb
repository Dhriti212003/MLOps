**Box plot**

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x='Outcome', y='Glucose', data=df)

plt.title('Glucose Levels by Diabetes Outcome')
plt.show()
```

**Bar graph**

```python
avg_blood_pressure = df.groupby('Outcome')['BloodPressure'].mean()

plt.bar(['No Diabetes', 'Diabetes'], avg_blood_pressure, color=['lightblue', 'lightcoral'])

plt.title('Average Blood Pressure by Diabetes Outcome')
plt.ylabel('Average Blood Pressure')

plt.show()
```

**Pie chart**

```python
outcome_counts = df['Outcome'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(outcome_counts, labels=['No Diabetes (0)', 'Diabetes (1)'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])

plt.title('Distribution of Diabetes Outcomes', fontsize=16)

plt.axis('equal')
plt.show()
```

**Scatter plot**

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['BloodPressure'], color='blue', alpha=0.5)

plt.title('Scatter Plot: Blood Pressure vs. Age', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Blood Pressure', fontsize=12)

plt.show()
```

**Histogram**

```python
plt.figure(figsize=(8, 6))
plt.hist(df['BloodPressure'], bins=15, color='lightgreen', edgecolor='black')

plt.title('Histogram of Blood Pressure', fontsize=16)
plt.xlabel('Blood Pressure', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
```

**Correlation matrix**

```python

corr_matrix = df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, square=True)

plt.title('Correlation Matrix of Diabetes Dataset', fontsize=16)

plt.tight_layout()
plt.show()
```
